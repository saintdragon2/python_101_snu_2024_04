```python
import pandas as pd

df = pd.read_csv('/Users/sungyonglee/github/python_101_snu_2024_04/hongong_ml_dl/s2024_0608/data/smoking_driking_dataset_Ver01.csv')
display(df)
```

```python
df = df.drop_duplicates()
```

```python
# 남녀 비율 출력
print(df['sex'].value_counts())

# 음주 여부 비율 출력
print(df['DRK_YN'].value_counts())

# 흡연 비율
print(df['SMK_stat_type_cd'].value_counts())
```

```python
# SMK_stat_type_cd 를 one-hot encoding

df = pd.get_dummies(df, columns=['SMK_stat_type_cd'], prefix='SMK', drop_first=True)
df = pd.get_dummies(df, columns=['sex'], prefix='sex', drop_first=True)
display(df)
```

```python
features = df.drop(columns=['DRK_YN']).to_numpy()
target = df['DRK_YN'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)

train_scaled = ss.transform(X_train)
test_scaled = ss.transform(X_test)
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(loss='hinge', random_state=42, max_iter=10, learning_rate='adaptive', eta0=0.01, tol=1e-2)
sgd.fit(train_scaled, y_train)

print(sgd.score(train_scaled, y_train)) 
print(sgd.score(test_scaled, y_test))
```

```python
sgd.partial_fit(train_scaled, y_train)
print(sgd.score(train_scaled, y_train))
print(sgd.score(test_scaled, y_test))
```

```python
import numpy as np

train_score = []
test_score = []

classes = np.unique(y_train)
```

```python
from tqdm import tqdm

for _ in tqdm(range(0, 100)):
    sgd.partial_fit(train_scaled, y_train, classes=classes)
    train_score.append(sgd.score(train_scaled, y_train))
    test_score.append(sgd.score(test_scaled, y_test))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score, label='train')
plt.plot(test_score, label='test')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.show()
```
